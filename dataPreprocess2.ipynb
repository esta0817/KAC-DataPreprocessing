```python
import pandas as pd
import os
import re
```

```python
### SID chart ###
df_path = "C:\\Users\esrikr\\GitHub\\RAT-Korea Airport Corporation\\KACData\\[첨부2]SID chart.xlsx"

### Set First DataFrame ###
df = pd.read_excel(df_path, sheet_name='Sheet1', header=2)
dropCol = list(range(3))
df = df.drop(df.columns[dropCol], axis=1)
df.head()
```

```python
coords = df["Coordinates"]

### Get latitude longitude from Coordinates column ###
Ncoords = []
Ecoords = []

for coord in coords:
    Nindex = coord.find("N")
    Eindex = coord.find("E")

    Nval = re.sub(r'[^0-9.]', '', coord[0:Nindex]) #N앞 숫자와 .만 추출
    Eval = re.sub(r'[^0-9.]', '', coord[Nindex+1:Eindex]) #N과 E사이 숫자와 .만 추출
    Ncoords.append(Nval)
    Ecoords.append(Eval)

### Create new columns to store coordinate ###
newdf = df
newdf["latitude"] = Ncoords
newdf["longitude"] = Ecoords

newdf.head()
```

```python
### Export to CSV ###
csv_path = df_path[0:-5]+"_new"+".csv"
newdf.to_csv(csv_path, sep=',')
```
